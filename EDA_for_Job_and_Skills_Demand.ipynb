**Data Cleaning:**                                               
Handle missing values using df.isnull().sum() to identify them, and then impute or remove them using df.fillna() or df.dropna()

```python
df.isnull().sum()
```

```python
df.fillna('Unknown')
```

**Clean Experience:**   
Convert values like "2 - 7 yrs" into numeric (average)

```python
def average_experience(exp):
    try:
        if pd.isnull(exp):
            return None
        parts = exp.replace("yrs", "").replace("yr", "").strip().split("-")
        return (float(parts[0]) + float(parts[1])) / 2
    except:
        return None

df['experience_avg'] = df['experience'].apply(average_experience)
```

**Clean Payrate:**   
Extract numeric salary from entries like "1,50,000 - 2,00,000"

```python
import re

def extract_salary(value):
    if pd.isna(value) or 'Not Disclos' or 'Unknown' in value:
        return None
    match = re.findall(r'\d[\d,]*', value)
    if match:
        return float(match[0].replace(',', ''))
    return None

df['salary_estimate'] = df['payrate'].apply(extract_salary)
```

**Clean Job Location:**  
Sometimes joblocation_address has multiple cities

```python
# Create one row per location if needed
df['joblocation_address'] = df['joblocation_address'].str.split(', ')
df = df.explode('joblocation_address')

# Reset the index after exploding to ensure it is unique
df = df.reset_index(drop=True)
```

**Save Cleaned Data**

```python
df.to_csv("cleaned_naukri_jobs.csv", index=False)
```

# **Visualizations for Job & Skills Demand**

**1.Top Job Locations**

```python
top_locs = df['joblocation_address'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_locs.values, y=top_locs.index, palette="viridis")
plt.title("Top 10 Job Locations")
plt.xlabel("Number of Jobs")
plt.ylabel("Location")
plt.show()

```

**2.Top Hiring Companies Bar Chart**

```python
top_companies = df['company'].value_counts().head(10)

top_companies.plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.title("Top 10 Hiring Companies")
plt.xlabel("Job Postings")
plt.ylabel("Company")
plt.gca().invert_yaxis()
plt.show()
```

**3.Most In-Demand Job Titles**

```python
top_titles = df['jobtitle'].value_counts().head(5)

plt.figure(figsize=(7, 7))
plt.pie(top_titles.values, labels=top_titles.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Top Job Titles Distribution")
plt.axis('equal')
plt.show()
```

**4.Most In-Demand Skills**

```python
# Extract skills from jobdescription
skills = ['python', 'java', 'sql', 'excel', 'aws', 'html', 'css', 'react', 'javascript', 'c++']
df['jobdescription'] = df['jobdescription'].astype(str).str.lower()

from collections import Counter
skill_counter = Counter()

for jd in df['jobdescription']:
    for skill in skills:
        if skill in jd:
            skill_counter[skill] += 1

skill_df = pd.DataFrame(skill_counter.items(), columns=['Skill', 'Count'])

import plotly.express as px
fig = px.treemap(skill_df, path=['Skill'], values='Count', title='Most In-Demand Skills')
fig.show()
```

**5.Salary Estimates**

```python
def extract_salary(value):
    if pd.isna(value) or 'Not Disclos' in value:
        return None
    match = re.findall(r'\d[\d,]*', value)
    if match:
        return float(match[0].replace(',', ''))
    return None

df['salary_estimate'] = df['payrate'].apply(extract_salary)

plt.figure(figsize=(8, 5))
sns.boxplot(y=df['salary_estimate'].dropna(), color='salmon')
plt.title("Salary Estimate Spread")
plt.ylabel("Salary (INR)")
plt.show()

```

**7.Joint Analysis: Experience vs Salary**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='experience_avg', y='salary_estimate', data=df, color='purple')
plt.title("Experience vs Salary")
plt.xlabel("Experience (Years)")
plt.ylabel("Salary Estimate (INR)")
plt.grid(True)
plt.show()
```

**8.Experience Requirement**

```python
def avg_exp(x):
    try:
        x = x.replace("yrs", "").replace("yr", "").strip().split('-')
        return (float(x[0]) + float(x[1])) / 2
    except:
        return None

df['experience_avg'] = df['experience'].apply(avg_exp)

plt.figure(figsize=(10, 5))
sns.histplot(df['experience_avg'].dropna(), bins=10, kde=True, color='teal')
plt.title("Experience Requirement Distribution")
plt.xlabel("Years of Experience")
plt.ylabel("Job Count")
plt.show()
```
